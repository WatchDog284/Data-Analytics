# Python Visualization Libraries

## Matplotlib

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create plot
plt.plot(x, y, color='green', marker='o', linestyle='dashed')
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend('x')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr'], 'Sales': [200, 300, 250, 400]}
df = pd.DataFrame(data)

# Line plot for sales trend
plt.plot(df['Month'], df['Sales'], marker='o', color='Red', linestyle='-.')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.savefig('Monthly Sales Trend.png')
plt.show()
```

## Seaborn - Advanced

Seaborn is a powerful Python library for creating advanced visualizations with ease. It builds on Matplotlib and provides a high-level interface for creating stunning statistical graphics. Here are some advanced plotting techniques you can try with Seaborn

### 1. pairplot() with customization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("iris")
print(data.columns)
print(data.dtypes)

sns.pairplot(data, hue='species', palette='coolwarm', diag_kind='kde')
plt.show()
```

### 2. Heatmap with Annotations

```python
#sns.heatmap(data.corr(), annot=True, fmt=".2f", cmap='YlGnBu', linewidths=0.5)
#plt.title("Correlation Heatmap")
#plt.show()

numeric_data = data.select_dtypes(include=["number"])  # Select numeric columns
sns.heatmap(numeric_data.corr(), annot=True, fmt=".2f", cmap='YlGnBu', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

data_numeric = data.drop(columns=['species'])  # Replace 'species' with any non-numeric column name
sns.heatmap(data_numeric.corr(), annot=True, fmt=".2f", cmap='YlGnBu', linewidths=0.5)
plt.show()
```

### 3. Custom Violin Plots

```python
#sns.violinplot(x='species', y='sepal_length', data=data, inner='quartile', palette='muted')
#plt.show()

sns.violinplot(x='species', y='sepal_length', hue='species', data=data, inner='quartile', palette='muted', legend=False)
plt.show()

data['value'] = data['sepal_length'] + data['sepal_width']  # Example transformation
sns.violinplot(x='species', y='value', hue='species', data=data, inner='quartile', palette='muted')
plt.show()
```

### 4. Regression plots with Multiple layers

```python
sns.lmplot(x='sepal_length', y='sepal_width', hue='species', data=data, aspect=1.5, markers=["o", "s", "^"])
plt.show()
```

### 5. FaceGrid for Subplots

```python
g = sns.FacetGrid(data, col="species", hue="species", col_wrap=4, palette="viridis")
g.map(plt.scatter, "sepal_length", "sepal_width", alpha=0.7)
g.add_legend()
plt.show()
```

### 6. Boxplot and Swarmplot Combo

```python
sns.boxplot(x='species', y='sepal_length', hue='species', data=data, palette='pastel', dodge=False, legend=False)
sns.swarmplot(x='species', y='sepal_length', data=data, color=".25")
plt.show()
```

### 7. Custom Style and Themes

```python
sns.set_theme(style="whitegrid", palette="bright")
sns.boxplot(x='species', y='sepal_length', data=data)
plt.show()
```
