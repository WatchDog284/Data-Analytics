# Data Normalization with Pandas

```python
import pandas as pd

df = pd.DataFrame([[180000, 110, 18.9, 1400],
                   [360000, 905, 23.4, 1800],
                   [230000, 230, 14.0, 1300],
                   [60000, 450, 13.5, 1500]],
                     
                    columns=['Col A', 'Col B', 'Col C', 'Col D']
                 )

display(df)
```

```python
import matplotlib.pyplot as plt

df.plot(kind='bar')
```

## Applying Normalization Technique for the above

### 1. Maximum absolute scaling

```python
df_max_scaled = df.copy()

for column in df_max_scaled.columns:
    df_max_scaled[column] = df_max_scaled[column] / df_max_scaled[column].abs().max()

display(df_max_scaled)
```

```python
import matplotlib.pyplot as plt

df_max_scaled.plot(kind='bar')
```

### 2. Using The min-max feature scaling / Normalization method

```python
df_min_max_scaled = df.copy()

for column in df_min_max_scaled.columns:
    df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())

print(df_min_max_scaled)
```

```python
import matplotlib.pyplot as plt

df_min_max_scaled.plot(kind = 'bar')
```

### 3. Using The z-score method / Standardization Method

```python
df_z_scaled = df.copy()

for column in df_z_scaled.columns:
    df_z_scaled[column] = (df_z_scaled[column] - df_z_scaled[column].mean()) / df_z_scaled[column].std()

display(df_z_scaled)
```

```python
import matplotlib.pyplot as plt

df_z_scaled.plot(kind = 'bar')
```
